```python
import pandas as pd
```

```python
df = pd.read_csv("results_df_SPY_1min.csv")
df
```

```python
df = df.sort_values("Cumulative Returns", ascending = False)
df
```

```python
metrics_df = df[["Reward", "Risk", "Profit Factor"]]
metrics_df
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Extract data from the DataFrame
x = metrics_df["Reward"]
y = metrics_df["Risk"]
z = metrics_df["Profit Factor"]

# Plot the data points as a scatter plot
ax.scatter(x, y, z, cmap='viridis', c=z)  # You can change the colormap

# Set labels for the axes
ax.set_xlabel('Reward')
ax.set_ylabel('Risk')
ax.set_zlabel('Profit Factor')
# Switch to an interactive backend
plt.switch_backend('tkagg')  # Use 'tkagg' backend for interactivity
plt.title('3D Scatter Plot of Reward, Risk, and Profit Factor')
plt.show()
```
